```python
# 各データ型に対する演算子の使用例
a = 10   # int
b = 3.5  # float
c = True  # bool
d = "hello"  # string
e = [1, 2, 3]  # list

# 算術演算子
print(a + 5)   # 整数の加算 15
print(b * 2)   # 浮動小数点の乗算 7.0
print(d + " world")  # 文字列の連結 hello world
print(e * 2)   # リストの繰り返し [1, 2, 3, 1, 2, 3]

# 比較演算子
print(a > 5)   # True
print(d == "hello")  # True
print(c == True)   # True

# 論理演算子
print(c and False)  # False
print(not c)   # False

# メンバーシップ演算子
print(2 in e)  # True
print("o" in d)  # True

# 同一性演算子
f = e
print(f is e)  # True
```
