```python
filename = "todo_list.txt"

try:
    with open(filename, "r", encoding="utf-8") as f:
        todos = [line.strip() for line in f]
except FileNotFoundError:
    todos = []

while True:
    num = input("番号を入力してください[1:追加、2:表示、3:削除、4:保存と終了]:")
    match num:
        case "1":
            todo = input("TODOを入力してください:")
            todos.append(todo)
        case "2":
            print("【TODOリスト】")
            for i, todo in enumerate(todos):
                print(f"{i}. {todo}")
        case "3":
            print("【TODOリスト】")
            for i, todo in enumerate(todos):
                print(f"{i}. {todo}")
                delete = input("削除する番号:")
            try:
                index = int(delete)
                todos.pop(index)
            except ValueError:
                print("数字を入力してください。")
            except IndexError:
                print("無効な番号です。")
        case "4":
            print("TODOリストを保存して終了します。")
            with open(filename, "w", encoding="utf-8") as file:
                file.writelines(f"{todo}\n" for todo in todos)
            break
        case _:
            print("不正な入力です")
            
```
