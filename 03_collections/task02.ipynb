```python
import string

words = []
dictionary = {}

while (word := input("英単語を入力してください：")) != "":
    words.append(word)

print(f"入力した英単語：{words}")

for character in string.ascii_lowercase:
    dictionary[character] = sum(s.count(character) for s in words)

for key, value in dictionary.items():
    if value > 0:
        print(f"{key}が{value}個ありました")
```
